# THIS is a preprocessing code Cause the features of our Dataset are slightly different from Previous one. 

```python
import pandas as pd
```

```python
data = pd.DataFrame({'original_image' : [], 'age' : [], 'gender' : []})
```

```python
data
```

```python
import os

# testing code for lite data set (20k)
# testing code for full data set

num = 1

for age in range (15,76) :
    # men
    path = 'AFAD-Full/'+str(age)+'/' + '111'
    file_list = os.listdir(path)
    gender = 'm'
    
    for x in file_list :
        data.loc[num] = [x,age,gender]
        num+=1
    # women
    path = 'AFAD-Full/'+str(age)+'/' + '112'
    file_list = os.listdir(path)
    gender = 'f'
    
    for x in file_list :
        data.loc[num] = [x,age,gender]
        num+=1
```

```python
data
```

```python
# 15-20,21- 25-32, 38-43, 48-53, 60-
# (15, 19) (20, 24) (25, 29) (30, 34) (35, 39)
# (40,44), (45, 49), (50, 59), (60, 100)

# Binning

data.loc[(15<=data['age']) & (data['age']<=19),'age(C)'] = '(18, 19)' # 청소년
data.loc[(20<=data['age']) & (data['age']<=24),'age(C)'] = '(20, 24)' # 20대 초반
data.loc[(25<=data['age']) & (data['age']<=29),'age(C)'] = '(25, 29)' # 20대 중후반
data.loc[(30<=data['age']) & (data['age']<=34),'age(C)'] = '(30, 34)' # 30대 초반
data.loc[(35<=data['age']) & (data['age']<=39),'age(C)'] = '(35, 39)' # 30대 중후반
data.loc[(40<=data['age']) & (data['age']<=44),'age(C)'] = '(40, 44)' # 40대 초반
data.loc[(45<=data['age']) & (data['age']<=49),'age(C)'] = '(45, 49)' # 40대 중후반
data.loc[(50<=data['age']) & (data['age']<=59),'age(C)'] = '(50, 59)' # 50대 
data.loc[(60<=data['age']) & (data['age']<=100),'age(C)'] = '(60, 100)' # 60대 이상
```
